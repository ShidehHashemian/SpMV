# Load Data Saved While Executatin of Different SpMSpv Multiplication in CPP

```python
# importing the pandas library
import pandas as pd

# reading the csv file using read_csv
# storing the data frame in variable called df
df = pd.read_csv('evaluation_data.csv')

data_val = {v_spf: {alg:{"time":list(), 'mspf': list()} for alg in set(df[:]["alg"])} for v_spf in sorted(list(set(df[:]["vspf"])))}
data_size = df.shape[0]
for i in range(data_size):
    data_val[df["vspf"][i]][df["alg"][i]]["time"].append(df["microsec"][i])
    data_val[df["vspf"][i]][df["alg"][i]]["mspf"].append(df["mspf"][i])

```

# Use Data to Plot Each Algorithm's Runtime By the Input Matrix Sparsity Factor For each Vector Sparsity Factor  

* Martix/Vector Sparsity Factor represent the probability of zero cells to number of cells
* Each algorithm's label name  is listed bellow
    1. coo (basic Matrix Format) 
    2. css (basic Matrix Format)
    3. csr (basic Matrix Format)
    4. ell (basic Matrix Format)
    5. coo-pll (COO format, implemented in parallel)
    6. csr-pll (CSR format, implemented in parallel)
    7. ell-pll (ELL format, implemented in parallel)
    8. csr-simd (CSR format, implemented using SIMD vectorization)

```python
import matplotlib.pyplot as plt

for vspf in data_val:
    for alg in data_val[vspf]:
        plt.plot(data_val[vspf][alg]["time"], data_val[vspf][alg]["mspf"], label = alg)
        plt.title("Vector Sparsity Factor: "+str(vspf))
        plt.xlabel("executation time (micro sec)")
        plt.ylabel("Matix Sparsity Factor")
    plt.legend()
    plt.show()
```
